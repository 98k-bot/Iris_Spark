```python
import os
import pandas as pd

data_path = "/common/zhangz2lab/protein_sequence_difference/Cardioboost_protein_sequences/"
input_df = pd.read_csv(os.path.join(data_path +'cm_train_protein_seq_df.csv'))
print(input_df.shape)
for i in range(input_df.shape[0]):
    str1 = input_df.loc[i, 'Mutated_Protein_Sequence']
    str2 = input_df.loc[i, 'Original_Protein_Sequence']
    difference_1 = "".join([char for char in str1 if char not in str2])
    difference_2 = "".join([char for char in str2 if char not in str1])

    print(f"Characters in str1 but not in str2: {difference_1}")
    print(f"Characters in str2 but not in str1: {difference_2}")

input_df = pd.read_csv(os.path.join(data_path +'cm_holdout_test_protein_seq_df.csv'))
print(input_df.shape)
for i in range(input_df.shape[0]):
    str1 = input_df.loc[i, 'Mutated_Protein_Sequence']
    str2 = input_df.loc[i, 'Original_Protein_Sequence']
    difference_1 = "".join([char for char in str1 if char not in str2])
    difference_2 = "".join([char for char in str2 if char not in str1])

    print(f"Characters in str1 but not in str2: {difference_1}")
    print(f"Characters in str2 but not in str1: {difference_2}")
```
